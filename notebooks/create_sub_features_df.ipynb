```python
import os
from utils.auxiliary_func import *
from utils.plots_func import *
import config.const as const
import config.params as params
import tifffile 
```

```python
DATA_PREPARATION_DIR = const.DATA_PREPARATION_DIR
```

### Core

```python
df_file_path = os.path.join(const.FULL_TISSUE_DIR, 'core', "FLIM_features_full_tissue.csv")
full_tissue_features_df_core = pd.read_csv(df_file_path, dtype = {'leap_ID': str})

full_tissue_features_df_core.head()
```

```python
# Filter and select relevant columns
sub_features_df_core = full_tissue_features_df_core[
    full_tissue_features_df_core['leap_ID'].isin(['043', '048'])  # keep only 043 and 048
][['leap_ID', 'nucleus_label', 'lifetime_mean', 'X coordinate', 'Y coordinate']]
```

### Resection

```python
df_file_path = os.path.join(const.FULL_TISSUE_DIR, 'resection', "FLIM_features_full_tissue.csv")
res_full_tissue_features_df = pd.read_csv(df_file_path, dtype = {'leap_ID': str})
res_full_tissue_features_df.head()
```

```python
# Filter and select relevant columns
sub_features_df_res = res_full_tissue_features_df[
    res_full_tissue_features_df['leap_ID'].isin(['049'])
][['leap_ID', 'nucleus_label', 'lifetime_mean', 'X coordinate', 'Y coordinate']]
```

### Merge between the resection sub df and core sub df

```python
combined_df = pd.concat([sub_features_df_core, sub_features_df_res], ignore_index=True)
```

```python
sub_features_path = os.path.join(DATA_PREPARATION_DIR, "sub_features_df.csv")
combined_df.to_csv(sub_features_path, index=False)
```
