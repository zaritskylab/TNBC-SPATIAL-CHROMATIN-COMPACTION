```python
from flim_analysis.feature_extraction.extract_features import *
import config.params as params
import config.const as const
from utils.auxiliary_func import *
import pandas as pd
```

```python
sample_type = 'core'
feature_file_name = "FLIM_features_full_tissue.csv"
```

```python
# Load data
df_file_path = os.path.join(const.FULL_TISSUE_DIR, sample_type, feature_file_name)
print(f"Reading FLIM feature file from: {df_file_path}")
full_tissue_features_df = pd.read_csv(df_file_path, dtype={'leap_ID': str})
print("Data loaded successfully.\n")

bins_amount = 18
```

```python
feature_name = 'area'
build_distribution(full_tissue_features_df, feature_name, bins_amount, sample_type)

max_val = full_tissue_features_df[feature_name].max()
bin_range = max_val / bins_amount
file_name_dist_full_tissue = f"features_{feature_name}_distribution_data_max_val_{max_val}_bins_amount_{bins_amount}_bin_range_{bin_range:.3f}.csv"
df_file_path = os.path.join(const.FULL_TISSUE_DIR, 'core', file_name_dist_full_tissue)
distribution_full_tissue_df = pd.read_csv(df_file_path, dtype = {'leap_ID': str})
distribution_full_tissue_df.head()
```

```python
feature_name = 'eccentricity'
build_distribution(full_tissue_features_df, feature_name, bins_amount, sample_type)

max_val = full_tissue_features_df[feature_name].max()
bin_range = max_val / bins_amount
file_name_dist_full_tissue = f"features_{feature_name}_distribution_data_max_val_{max_val}_bins_amount_{bins_amount}_bin_range_{bin_range:.3f}.csv"
df_file_path = os.path.join(const.FULL_TISSUE_DIR, 'core', file_name_dist_full_tissue)
distribution_full_tissue_df = pd.read_csv(df_file_path, dtype = {'leap_ID': str})
distribution_full_tissue_df.head()
```
