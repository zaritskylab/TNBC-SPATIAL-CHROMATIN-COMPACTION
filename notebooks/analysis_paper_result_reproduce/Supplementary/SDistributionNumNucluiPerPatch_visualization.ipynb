## Define Parameters

```python
patch_size = 1500
overlap = 0.75
```

```python
data_preparation_dir = os.path.join(const.base_dir, 'figure_results', 'data_preparation')

figure_supplementary_dir = os.path.join(const.base_dir, 'figure_results', 'Supplementary')
os.makedirs(figure_supplementary_dir, exist_ok=True)
```

## Plot distribution of nuclei amount per patch

```python
nuclei_patch_path = os.path.join(data_preparation_dir, f'nuclei_count_per_patch_{patch_size}_df.csv')
```

```python
nuclei_count_per_patch = pd.read_csv(nuclei_patch_path, dtype = {'leap_ID': str})
```

```python
# Print the max and min
print(f"Max nuclei per patch: {nuclei_count_per_patch.max()}")
print(f"Min nuclei per patch: {nuclei_count_per_patch.min()}")

# Plot the distribution
plt.figure(figsize=(8, 6))
plt.hist(nuclei_count_per_patch['nuclei_count'], bins=100, edgecolor='black')
plt.title('Distribution of Number of Nuclei per Patch', fontsize=16, fontweight='bold')
plt.xlabel('Number of Nuclei per Patch', fontsize=14)
plt.ylabel('Count of Patches', fontsize=14)


save_file_name= f"SDistributionNumNucluiPerPatch_{patch_size}"
save_fig(figure_supplementary_dir, save_file_name, 'pdf', plt)
```
