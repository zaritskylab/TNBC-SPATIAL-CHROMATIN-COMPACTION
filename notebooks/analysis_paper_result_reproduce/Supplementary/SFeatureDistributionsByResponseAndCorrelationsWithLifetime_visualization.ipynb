# Feature Distributions By Response And Features Correlations With Lifetime

### Define Parameters

```python
data_preparation_dir = os.path.join(const.base_dir, 'figure_results', 'data_preparation')


figure_supplementary_dir = os.path.join(const.base_dir, 'figure_results', 'Supplementary')
os.makedirs(figure_supplementary_dir, exist_ok=True)
```

### Lifetime correltion with another feature

```python
df_file_path = os.path.join(data_preparation_dir, 'correlation_df_between_lifetime_and_morphological_density_features.csv')


df_correlation = pd.read_csv(df_file_path, dtype = {'leap_ID': str})
```

```python
plt_fig = plot_boxplot_by_category(df_correlation, 'Correlation', category_col='feature_name', figsize=(8,4))
save_fig(figure_supplementary_dir, 'correlation_by_feature_name', 'pdf', plt_fig)
```

### Responder vs Non responder seperation by feature

```python
df_median_file_path = os.path.join(data_preparation_dir, "features_median_data_core.csv")
median_df = pd.read_csv(df_median_file_path, dtype={'leap_ID': str})
median_df.head(2)
```

```python
plt_fig = category_seperation_by_feature(median_df, 'eccentricity', order = ['responder', 'non responder'])
save_fig(figure_supplementary_dir, 'eccentricity_reponder_non_responder', 'pdf', plt_fig)
```

```python
plt_fig = category_seperation_by_feature(median_df, 'area', mapping={'non responder': 0, 'responder': 1}, order = ['responder', 'non responder'])
save_fig(figure_supplementary_dir, 'area_reponder_non_responder', 'pdf', plt_fig)
```

```python
plt_fig = category_seperation_by_feature(median_df, 'density_radius_40', mapping={'non responder': 0, 'responder': 1}, order = ['responder', 'non responder'])
save_fig(figure_supplementary_dir, 'density_reponder_non_responder', 'pdf', plt_fig)
```
